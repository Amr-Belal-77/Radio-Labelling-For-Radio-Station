```python

def Adjacency(status,X_num):
    
    A=[[0]*X_num for i in range(X_num)]             
    if status ==1 :                                               # --------- >> open << ---------
        for i, row in enumerate(A):
            
            for j, value in enumerate(row):
                if i == j:
                    if i and j != 0  and i and j !=len(A)-1 :
                        A[i][j-1]=1
                        A[i][j+1]=1
                    
                    if i == j and i==0  :
                        A[i][j+1]=1

                    if i == j and i==len(A)-1  :
                        A[i][j]=1
    # return A
        # print('A = ')
        # for row in A :
        #     print(row)

        # ******************************************************************************************************************            

    elif status == 2:                                             # --------- >> cycle << --------
        for i, row in enumerate(A):
            for j, value in enumerate(row):

                if i == j:
                    if i and j != 0  and i and j !=len(A)-1 :
                        A[i][j-1]=1
                        A[i][j+1]=1
                    
                    if  i==0  :                     # first row
                        A[i][j+1]=1
                        A[i][len(A)-1]=1 

                    if i == j and i==len(A)-1  :    #   last row
                        A[i][j]=1
                        A[i][0]=1   
        print('A = ')
        for row in A :
            print(row)

        # ******************************************************************************************************************

    elif status ==3 :                                             # --------- >> complete << ---------
        for i, row in enumerate(A):
            for j, value in enumerate(row):
                    if i !=j   :
                        A[i][j]=1
        print('A = ')
        for row in A :
            print(row)
    return A
A=Adjacency(status,X_num)
print(A)
```
